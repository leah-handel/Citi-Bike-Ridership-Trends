```python
import pandas as pd
import requests
import json
import time
```

```python
# reading in data

nyc_apr21 = pd.read_csv("NYdata/202104-citibike-tripdata.csv")
nyc_apr20 = pd.read_csv("NYdata/202004-citibike-tripdata.csv")
nyc_apr19 = pd.read_csv("NYdata/201904-citibike-tripdata.csv")
```

```python
nyc_data = pd.concat([nyc_apr21, nyc_apr20, nyc_apr19])
```

```python
nyc_data
```

```python
jc_apr21 = pd.read_csv("JCdata/JC-202104-citibike-tripdata.csv")
jc_apr20 = pd.read_csv("JCdata/JC-202004-citibike-tripdata.csv")
jc_apr19 = pd.read_csv("JCdata/JC-201904-citibike-tripdata.csv")
```

```python
jc_data = pd.concat([jc_apr19, jc_apr20, jc_apr21])
```

```python
data = pd.concat([nyc_data, jc_data])
```

```python
# reverse geocoding zipcode and borough, inspired by a similar project: 
# https://towardsdatascience.com/reverse-geocoding-with-nyc-bike-share-data-cdef427987f8
# to start, making a dataframe of all the stations involved.

start_stations = data[['start station id','start station name',
                       'start station latitude','start station longitude']]
start_stations = start_stations.rename(columns={'start station id':'id',
                               'start station name':'name',
                               'start station latitude':'latitude',
                               'start station longitude':'longitude'})

```

```python
end_stations = data[['end station id','end station name','end station latitude','end station longitude']]

end_stations = end_stations.rename(columns={'end station id':'id',
                               'end station name':'name',
                               'end station latitude':'latitude',
                               'end station longitude':'longitude'})
```

```python
all_stations = pd.concat([start_stations, end_stations]).drop_duplicates(subset = ['id'])
```

```python
all_stations
```

```python
test_stations = all_stations.loc[1:10]
```

```python
# testing the api call

for index, row in test_stations.iterrows():
    
    try:
        time.sleep(1)
        
        lat = row["latitude"]
        long = row["longitude"]
        station = row["id"]
    
        response = requests.get(f"https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={long}&format=json&email=leah.handel@gmail.com&zoom=18").json()
    
        test_stations.loc[index, "borough"] = response["address"]["suburb"]
        test_stations.loc[index, "zipcode"] = response["address"]["postcode"]
        
        print(f"got station {station}")
    
    except:
        print(f"error with station {station}")
```

```python
test_stations
```

```python
# the whole API call, warning that this is a long one!

for index, row in all_stations.iterrows():
    
    try:
        time.sleep(1)
        
        lat = row["latitude"]
        long = row["longitude"]
        station = row["id"]
    
        response = requests.get(f"https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={long}&format=json&email=leah.handel@gmail.com&zoom=18").json()
    
        all_stations.loc[index, "borough"] = response["address"]["suburb"]
        all_stations.loc[index, "zipcode"] = response["address"]["postcode"]
        
        print(f"got station {station}")
    
    except:
        print(f"error with station {station}")
```

```python
# Of course it turned out that grabbing the borough for stations in jersey city didn't work!

all_stations.loc[pd.isnull(all_stations["zipcode"]),:]
```

```python
requests.get(f"https://nominatim.openstreetmap.org/reverse?lat=40.724605&lon=-74.078406&format=json&email=leah.handel@gmail.com&zoom=18").json()
```

```python
# Redoing the api call for the stations it missed the first time. 
# Could have just manually filled in Jersey City but we were also missing zip code.

for index, row in all_stations.iterrows():
    
    if pd.isnull(row["zipcode"]):
    
        try:
            time.sleep(1)
        
            lat = row["latitude"]
            long = row["longitude"]
            station = row["id"]
        
            response = requests.get(f"https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={long}&format=json&email=leah.handel@gmail.com&zoom=18").json()
        
            all_stations.loc[index, "borough"] = response["address"]["city"]
            all_stations.loc[index, "zipcode"] = response["address"]["postcode"]
            
            print(f"got station {station}")
        
        except:
            print(f"error with station {station}")
```

```python
# checking for weird zip values and cleaning

all_stations.loc[all_stations["zipcode"].str.len()!=5,:]
```

```python
all_stations[['zipcode','extra']] = all_stations.zipcode.str.split("-",expand=True)
```

```python
all_stations = all_stations.drop(columns="extra")
```

```python
all_stations.loc[all_stations["zipcode"].str.len()!=5,:]
```

```python
all_stations["zipcode"] = all_stations["zipcode"].replace(["10451:10455"], ["10451"])
all_stations["zipcode"] = all_stations["zipcode"].replace(["112321"], ["11231"])
all_stations["zipcode"] = all_stations["zipcode"].replace(["100036"], ["10036"])
all_stations["zipcode"] = all_stations["zipcode"].replace(["100014"], ["10014"])
all_stations["zipcode"] = all_stations["zipcode"].replace(["100"], ["10023"])
```

```python
all_stations.loc[all_stations["zipcode"].str.len()!=5,:]
```

```python
merge_stations = all_stations[["id", "borough", "zipcode"]]
```

```python
#merging zip and borough data into the ride data

data = data.merge(merge_stations, how="left", left_on="start station id", right_on="id", suffixes=(None, "_start"))
```

```python
data.head()
```

```python
data = data.rename(columns={"id":"start id","borough":"start borough","zipcode":"start zipcode"})
```

```python
station_data = data[['tripduration', 'starttime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',"start borough","start zipcode"]]
```

```python
# A few NYC stations got caught in our second API call and had their borough listed as New York

station_data["start borough"].unique()
```

```python
station_data["start borough"] = station_data["start borough"].replace(["New York"], ["Manhattan"])
```

```python
# export for tableau!
station_data.to_csv("aprils.csv", index=False)
```
